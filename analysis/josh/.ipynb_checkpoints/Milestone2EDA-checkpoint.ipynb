# EDA: Exploring Weapon Usage in Homicides over the Years
Here I am going to conduct some visual analysis and exploration of the frequency and usage of weapons in the data
### First let's prepare the dataset

```python
import sys, os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sys.path.insert(0,os.path.abspath('..'))
from Scripts.project_functions import *
sns.set_theme(style="ticks", font_scale=1.3,)
plt.rc("axes.spines", top=False, right=False)
```

```python
df = load_and_process('...data.raw.database')
```

### And now let's clean it up for our uses specifically

```python
#Many are the data type object which is hard to utilize
df = df.convert_dtypes()
#For our purposes we will drop rows with an Unknown weapon as that doesn't assist our purposes too well
df = df.loc[lambda x: x['Weapon'] != 'Unknown']
print(df.nunique())
```

### We can see from the unique count there are 15 unique weapons used over 35 years. This is the data we will be exploring.

#### Let's begin by constructing a simple histogram plot with the weapon count data

```python
plt.figure(figsize=(14,9))
dplot = sns.displot(df['Weapon'], kde=False, height = 5, aspect = 4)
dplot.set(title = 'Unique Weapon Count', xlabel = 'Weapon Type')
```

### Finally let's analyze the heatmap of the count of unique weapons used over the years

```python
#First create our dataset with grouping by year and weapon
df2 =(
    df.groupby(['Year', 'Weapon']).size()
        .to_frame(name = 'murder_count').reset_index()
        .loc[lambda x: x['Weapon'] != 'Unknown']
        .reset_index(drop=True)
)

df2 = df2.pivot_table(index='Year', columns='Weapon', values='murder_count')
# You can separate data with lines
sns.heatmap(df2, cmap='Blues', linecolor='white')
```

We find that handgun killings were very popular around the 1992-1995 years but are vastly the most popular method of homiecide
Other than that we can see that knife killings and blunt object killings are around the next most popular areas with blunt objects being very popular
### We can see how prevalent handgun killings are. How about we remove it to see any findings?

```python
#First create our dataset with grouping by year and weapon
df3 =(
    df.groupby(['Year', 'Weapon']).size()
        .to_frame(name = 'murder_count').reset_index()
        .loc[lambda x: x['Weapon'] != 'Unknown']
        .loc[lambda x: x['Weapon'] != 'Handgun']
        .reset_index(drop=True)
)

df3 = df3.pivot_table(index='Year', columns='Weapon', values='murder_count')
# You can separate data with lines
sns.heatmap(df3, cmap='Blues', linecolor='white')
```

### Here we see that Knife takes over as the most popular method
Knifes killing have declined since the 1980's and is less popular in in the 00's
This is 'replaced' by Firearm killings which have gained popularity since the 1980's
Along with this we can observe that Shotgun and blunt object killings have declined in popularity since the 1980's

### Lastly let us observe a violin plot of the data to confirm our observations

```python
df4 =(
    df.groupby(['Year', 'Weapon']).size()
        .to_frame(name = 'murder_count').reset_index()
        .loc[lambda x: x['Weapon'] != 'Unknown']
        .reset_index(drop=True)
)
fig_dims = (22, 5)
fig, ax = plt.subplots(figsize=fig_dims)
sns.stripplot(x='Year',y='murder_count',data=df4, hue = 'Weapon', ax=ax)
plt.legend(loc=0)
```
